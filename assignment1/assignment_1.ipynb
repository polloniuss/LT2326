```python
# We create a dictionary with images as key and their list of polygon as values
# Example : {0000172:[[[140,896],[162,898],[162,935],[140,935]],[polygon,polygon,polygon,polygon],[...]]}

def get_bounding_box(data):
    with open('/scratch/lt2326-h21/a1/train.jsonl') as jsonl_f:
        jsonl_data = [json.loads(x) for x in jsonl_f]
        images_polygons = {}

        for dictionary in jsonl_data:
            if dictionary['file_name'] in data:

                for sentence in dictionary['annotations']:
                    for character in sentence:
                        if character['is_chinese']:
                            
                            #to avoid KeyError
                            if dictionary['file_name'] not in images_polygons:
                                images_polygons[dictionary['file_name']] = []
                                
                            images_polygons[dictionary['file_name']].append(character['polygon'])
        
        return images_polygons
```

```python
# We create a grid with the width and height that we found earlier
grid = np.array([[[a,b] for b in list(range(2048))] for a in list(range(2048))])
# We reverse the order of elements along axis 1 (left/right)
grid = np.rot90(np.fliplr(grid))
grid = grid.reshape(2048 * 2048, 2)

print(grid)
```

```python
# image to tensors to train
# tensors to np array to process
# binary class (2048 * 2048) = 4194304
# where a pixel is either in a chinese character bounding box or not
# gold / truth / expected values

"""
# https://datacarpentry.org/image-processing/aio/index.html
# coordinate system, x y
Imagine that we have a fairly large, but very boring image: a 5,000 × 5,000 pixel image composed of nothing but white pixels. If we used an uncompressed image format such as BMP, with the 24-bit RGB color model, how much storage would be required for the file?
In such an image, there are 5,000 × 5,000 = 25,000,000 pixels, and 24 bits for each pixel, leading to 25,000,000 × 24 = 600,000,000 bits, or 75,000,000 bytes (71.5MB). That is quite a lot of space for a very uninteresting image!

get images resized and tensored
"""
def get_truth(dataset,key):
    
    # image_polygon = np.zeros(2048*2048)
    image_polygon = np.zeros((2048,2048))
    
    # The images are too big and would require too much memory
    # So we resize them to 200 * 200
    image = mpimg.imread(path + key)
    image = transform.resize(image, (200, 200))
    
    # We go through each set of polygon points of the current image
    for polygon_point in dataset[key]:
        
        # We create a numpy array of our bounding box
        # with true when a pixel point is inside a bounding box
        polygon_path = mplpath.Path(polygon_point)
        polygon_grid = polygon_path.contains_points(grid)
        
        # We switch to integer values   
        truthvals = np.asarray(polygon_grid, int)
        
        # Correct dimension
        truthvals.shape = (2048, 2048)
        image_polygon += truthvals
    
    return image, image_polygon
```

```python
# Parallelize python code to save time on data processing
# https://stackoverflow.com/questions/42220458/what-does-the-delayed-function-do-when-used-with-joblib-in-python
# https://www.tutorialdocs.com/tutorial/joblib/examples.html

def process_in_parallel(dataset):

    tic = time.time()
    print("Curently processing data")
    processed_data = Parallel(n_jobs=10)(delayed(get_truth)(dataset, key) for key in list(dataset.keys()))    
    toc = time.time()

    print('Elapsed time for the entire processing: {:.2f} s'.format(toc - tic))
    
    return processed_data
```

```python
# representation of the images as training input
# representation of the bounding boxes as objective -> binary classification

# save the models for later
# output : soft binary = 0 or 1 if in a bounding box or not

# https://towardsdatascience.com/build-your-own-convolution-neural-network-in-5-mins-4217c2cf964f

# https://d2l.ai/chapter_computer-vision/object-detection-dataset.html
# https://thegradient.pub/semantic-segmentation/
```

```python
# From Amelie:
# https://www.pyimagesearch.com/2021/07/19/pytorch-training-your-first-convolutional-neural-network-cnn/

# https://d2l.ai/chapter_convolutional-modern/alexnet.html
# AlexNet model - Deep Concolutional Neural Networks

# https://d2l.ai/chapter_convolutional-neural-networks/lenet.html
# LeNet model - CNN

"""
Sigmoid at output is more suitable for binary classification. (probably suitable for multi-label classification as well).
Softmax at output is more suitable for multi-class classification
"""
#https://medium.com/dejunhuang/learning-day-57-practical-5-loss-function-crossentropyloss-vs-bceloss-in-pytorch-softmax-vs-bd866c8a0d23

class LeNet(nn.Module):

    def __init__(self): #(self, numChannels, classes)
        # call the parent constructor
        super(LeNet, self).__init__()
        
        # initialize first set of CONV => RELU => POOL layers
        # We have 3 channels because the pixels are colored
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=6, kernel_size=(3, 3), padding=(1, 1))
        self.relu1 = nn.ReLU()
        self.maxpool1 = nn.MaxPool2d(kernel_size=(4, 4))
        
        # initialize second set of CONV => RELU => POOL layers
        self.conv2 = nn.Conv2d(6, 12, (3, 3), stride=1, padding=(1, 1))
        self.relu2 = nn.ReLU()
        self.maxpool2 = nn.MaxPool2d(kernel_size=(2, 2))
        
        # initialize third set of CONV => RELU => POOL layers
        self.conv3 = nn.Conv2d(12,  24, (3, 3), stride=1, padding=(1, 1))
        self.relu3 = nn.ReLU()
        self.maxpool2 = nn.MaxPool2d(kernel_size=(2, 2))

        # Expected shape compatible with [4, 3456]
        self.fc1 = nn.Linear(24*12*12, 400)
        self.relu4 = nn.ReLU()
        
        self.fc2 = nn.Linear(in_features=400, out_features=200)
        self.sig = nn.Sigmoid()
        self.up = nn.Upsample((2048,2048))
        
    def forward(self, x):
        
        x = x.float().permute(0, 3, 1, 2)
        
        # pass the input through our first set
        # CONV => RELU => POOL layers
        x = self.conv1(x)
        x = self.relu1(x)
        x = self.maxpool1(x)
        
        # pass the output from the previous layer through the second set
        # CONV => RELU => POOL layers
        x = self.conv2(x)
        x = self.relu2(x)
        x = self.maxpool2(x)
        
        # pass the output from the previous layer through the third set
        # CONV => RELU => POOL layers
        x = self.conv3(x)
        x = self.relu3(x)
        x = self.maxpool2(x)
        
        x = self.fc1(torch.flatten(x, 1))
        x = self.relu4(x)
        
        # pass the output to our sigmoid classifier to get our output
        # predictions
        x = self.fc2(x)
        x = self.sig(x)        
        
        x = x.reshape(x.shape[0],1,200,1)
        x = self.up(x)
        
        output = x.permute(0, 2, 3, 1) 
        
        # return the output predictions
        return output
```

## Model 2

```python
# https://d2l.ai/chapter_convolutional-modern/batch-norm.html
# https://www.guru99.com/convnet-tensorflow-image-classification.html
# https://towardsdatascience.com/using-tensorflow-object-detection-to-do-pixel-wise-classification-702bf2605182

def batch_norm(X, gamma, beta, moving_mean, moving_var, eps, momentum):
    # Use `is_grad_enabled` to determine whether the current mode is training
    # mode or prediction mode
    if not torch.is_grad_enabled():
        # If it is prediction mode, directly use the mean and variance
        # obtained by moving average
        X_hat = (X - moving_mean) / torch.sqrt(moving_var + eps)
    else:
        assert len(X.shape) in (2, 4)
        if len(X.shape) == 2:
            # When using a fully-connected layer, calculate the mean and
            # variance on the feature dimension
            mean = X.mean(dim=0)
            var = ((X - mean) ** 2).mean(dim=0)
        else:
            # When using a two-dimensional convolutional layer, calculate the
            # mean and variance on the channel dimension (axis=1). Here we
            # need to maintain the shape of `X`, so that the broadcasting
            # operation can be carried out later
            mean = X.mean(dim=(0, 2, 3), keepdim=True)
            var = ((X - mean) ** 2).mean(dim=(0, 2, 3), keepdim=True)
        # In training mode, the current mean and variance are used for the
        # standardization
        X_hat = (X - mean) / torch.sqrt(var + eps)
        # Update the mean and variance using moving average
        moving_mean = momentum * moving_mean + (1.0 - momentum) * mean
        moving_var = momentum * moving_var + (1.0 - momentum) * var
    Y = gamma * X_hat + beta  # Scale and shift
    return Y, moving_mean.data, moving_var.data


class BatchNorm(nn.Module):
    # `num_features`: the number of outputs for a fully-connected layer
    # or the number of output channels for a convolutional layer. `num_dims`:
    # 2 for a fully-connected layer and 4 for a convolutional layer
    def __init__(self, num_features, num_dims):
        super().__init__()
        if num_dims == 2:
            shape = (1, num_features)
        else:
            shape = (1, num_features, 1, 1)
        # The scale parameter and the shift parameter (model parameters) are
        # initialized to 1 and 0, respectively
        self.gamma = nn.Parameter(torch.ones(shape))
        self.beta = nn.Parameter(torch.zeros(shape))
        # The variables that are not model parameters are initialized to 0 and 1
        self.moving_mean = torch.zeros(shape)
        self.moving_var = torch.ones(shape)

    def forward(self, X):
        # If `X` is not on the main memory, copy `moving_mean` and
        # `moving_var` to the device where `X` is located
        if self.moving_mean.device != X.device:
            self.moving_mean = self.moving_mean.to(X.device)
            self.moving_var = self.moving_var.to(X.device)
        # Save the updated `moving_mean` and `moving_var`
        Y, self.moving_mean, self.moving_var = batch_norm(
            X, self.gamma, self.beta, self.moving_mean,
            self.moving_var, eps=1e-5, momentum=0.9)
        return Y

class BatchLeNet(nn.Module):

    def __init__(self): #(self, numChannels, classes)
        # call the parent constructor
        super(BatchLeNet, self).__init__()
        
        # initialize first set of CONV => RELU => POOL layers
        # We have 3 channels because the pixels are colored
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=6, kernel_size=(3, 3), padding=(1, 1))
        self.batch1 = BatchNorm(6, num_dims=4)
        self.sig = nn.Sigmoid()
        self.avgpool1 = nn.AvgPool2d(kernel_size=(4, 4))
        
        # initialize second set of CONV => RELU => POOL layers
        self.conv2 = nn.Conv2d(6, 12, (3, 3), stride=1, padding=(1, 1))
        self.batch2 = BatchNorm(12, num_dims=4)
        self.sig = nn.Sigmoid()
        self.avgpool2 = nn.AvgPool2d(kernel_size=(2, 2))
        
        # initialize third set of CONV => RELU => POOL layers
        self.conv3 = nn.Conv2d(12,  24, (3, 3), stride=1, padding=(1, 1))
        self.batch3 = BatchNorm(24, num_dims=4)
        self.sig = nn.Sigmoid()
        self.avgpool2 = nn.AvgPool2d(kernel_size=(2, 2))
        self.flat = nn.Flatten()

        # Expected shape compatible with [4, 3456]
        self.fc1 = nn.Linear(24*12*12, 400)
        self.batch4 = BatchNorm(400, num_dims=2)
        self.sig = nn.Sigmoid()
        
        self.fc2 = nn.Linear(in_features=400, out_features=200)
        self.batch5 = BatchNorm(200, num_dims=2)
        self.sig = nn.Sigmoid()
        self.up = nn.Upsample((2048,2048))
        
    def forward(self, x):
        
        x = x.float().permute(0, 3, 1, 2)
        
        # pass the input through our first set
        # CONV => RELU => POOL layers
        x = self.conv1(x)
        x = self.batch1(x)
        x = self.sig(x)
        x = self.avgpool1(x)
        
        # pass the output from the previous layer through the second set
        # CONV => RELU => POOL layers
        x = self.conv2(x)
        x = self.batch2(x)
        x = self.sig(x)
        x = self.avgpool2(x)
        
        # pass the output from the previous layer through the third set
        # CONV => RELU => POOL layers
        x = self.conv3(x)
        x = self.batch3(x)
        x = self.sig(x)
        x = self.avgpool2(x)
        x = self.flat(x)
        
        x = self.fc1(torch.flatten(x, 1))
        x = self.batch4(x)
        x = self.sig(x)
        
        # pass the output to our sigmoid classifier to get our output
        # predictions
        x = self.fc2(x)
        x = self.batch5(x)
        x = self.sig(x)
        
        x = x.reshape(x.shape[0],1,200,1)
        x = self.up(x)
        
        output = x.permute(0, 2, 3, 1) 
        
        # return the output predictions
        return output
```

```python
# https://scikit-learn.org/stable/modules/classes.html#module-sklearn.metrics

def test_model(model, dataloader):
    
    # Mean squared error
    loss_fn = nn.MSELoss()
    #correct_pred = 0
    total_loss = 0

    # switch off autograd for evaluation
    with torch.no_grad():
        # set the model in evaluation mode
        model.eval()
        # loop over the validation set
        for i, (x, y) in enumerate(dataloader):
            # send the input to the device
            (x, y) = (x.to(device), y.to(device))
            # make the predictions and calculate the validation loss
            output = model(x)
            loss = loss_fn(output.reshape(-1), y.reshape(-1))
            total_loss += loss.item()
            # calculate the number of correct predictions
            # correct_pred += (output.argmax(1) == y).type(torch.float).sum().item()
            
            y_true = output.reshape(-1).cpu()
            y_pred = y.reshape(-1).cpu()
            
            # Mean squared error
            mse = mean_squared_error(y_true, y_pred)
            
    print("Total mean squared error: ", mse)
    print("Total loss: ", total_loss)
```

```python
def show_model(model, dataloader):
    
    pred_img = []

    with torch.no_grad():
        
        model.eval()
        
        for i, (x, y) in enumerate(dataloader):
            
            (x, y) = (x.to(device), y.to(device))
            output = model(x)
            pred_truth = torch.sigmoid(output)
            
            a = x.detach().cpu().numpy()
            b = y.detach().cpu().numpy()
            c = pred_truth.detach().cpu().numpy().squeeze(3)
            
            pred_img += list(zip(*[a,b,c]))
            
    img_example = random.choice(range(0, 168))

    for image in pred_img[img_example]:

        plt.figure()
        plt.imshow(image)
```

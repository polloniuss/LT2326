```python
def test_features(trnX, trny, tstX, tsty):
    accuracies = {}
    for col in trnX.columns:
        print("Testing by removing {}...".format(col))
        trnX2 = trnX.drop(col, axis=1)
        tstX2 = tstX.drop(col, axis=1)
        
        svm = SVC()
        
        svm.fit(trnX2, trny)
        preds = svm.predict(tstX2)
        
        acc = len(tstX[preds == tsty]) / len(tsty)
        
        print("Removing column {} yields accuracy {}.".format(col, acc))
        
        accuracies[col] = acc
    
    return accuracies
```

```python
accs = test_features(train_X, train_y, test_X, test_y)
```

```python
def accuracies_by_class(tstdf, tsty, preds):
    accuracies = {}
    tstdf2 = tstdf.copy()
    tstdf2['Class'] = tstdf2['Class'].apply(lambda x: str(x))
    correct = tstdf2[tsty == preds]
    for cls in tstdf2["Class"].unique():
        accuracies[cls] = (len(correct[correct['Class'] == cls])) / (len(tstdf2[tstdf2["Class"] == cls]))
        
    return accuracies
```

```python
accs = accuracies_by_class(testdf, test_y, predictions)
accs
```
